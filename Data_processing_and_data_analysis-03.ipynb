```python
#Problem Statement 1: Employee Performance Bonus Eligibility
```

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
highest = max(employees.values())
topper = []
for name, score in employees.items():
    if score == highest:
        topper.append(name)
print(f"Top Performers Eligible for Bonus: {', '.join(topper)} (Score: {highest})")
```

```python
#Problem Statement 2: Search Query Keyword Analysis
```

```python
query = "Buy mobile phone buy phone online"
query = query.lower().split()
count = {}
for que in query:
    if que in count:
        count[que] = count[que] + 1
    else:
        count[que] = 1
for que in count:
    if count[que] > 1:
        print(que, ":", count[que])
```

```python
#Problem Statement 3: Sensor Data Validation
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid = []
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid.append((index, value))
print("Valid Sensor Readings (Hour, Value):")
print(valid)
```

```python
#Problem Statement 4: Email Domain Usage Analysis
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domains = {}
total = len(emails)
for email in emails:
    domain = email.split('@')[1]
    domains[domain] = domains.get(domain, 0) + 1
for domain, count in domains.items():
    percentage = (count / total) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
#Problem Statement 5: Sales Spike Detection
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg = sum(sales) / len(sales)
threshold = avg * 1.3
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
#Problem Statement 6: Duplicate User ID Detection
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
count = {}
for user in user_ids:
    count[user] = count.get(user, 0) + 1
for user, freq in count.items():
    if freq > 1:
        print(f"{user} → {freq} times")
```
