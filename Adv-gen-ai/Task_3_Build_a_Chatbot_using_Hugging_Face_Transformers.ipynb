```python
import os
import warnings
import torch
from transformers import pipeline, logging

# ==========================================
# Suppressing Specific Warnings
# ==========================================
os.environ['HF_HUB_DISABLE_SYMLINKS_WARNING'] = '1'
warnings.filterwarnings('ignore')
logging.set_verbosity_error()

def run_chatbot():
    print("Loading transformer model... Please wait.")
    
    # Initialize a text-generation pipeline using GPT-2 
    # Pad token is set to EOS token (50256 for GPT-2) to avoid warnings
    generator = pipeline('text-generation', model='gpt2', pad_token_id=50256)
    
    print("\nChatbot: Hello! I am your AI assistant. How can I help you today?")
    
    # Continuous interaction loop
    while True:
        # Accept user input from the console
        user_input = input("User: ")
        
        # Exit Condition: Stop chatbot when user types 'exit' or 'quit'
        if user_input.lower() in ['exit', 'quit']:
            print("Chatbot ends the conversation.")
            break
            
        # ==========================================
        # Prompt Engineering (Few-Shot Learning)
        # ==========================================
        
        prompt = f"""You are a helpful and intelligent AI assistant.
User: Hello
Chatbot: Hello! Nice to meet you. How can I assist you today?
User: What is Artificial Intelligence?
Chatbot: Artificial Intelligence refers to the simulation of human intelligence by machines that can perform tasks such as learning, reasoning, and problem solving.
User: Who created Python?
Chatbot: Python was created by Guido van Rossum and released in 1991.
User: Thank you
Chatbot: You're welcome! Feel free to ask more questions.
User: {user_input}
Chatbot:"""

        # Generate the response
        result = generator(
            prompt, 
            max_new_tokens=40,    # Limit generation length
            num_return_sequences=1, 
            temperature=0.5,      # Lower temperature = more predictable, structured responses
            do_sample=True,
            truncation=True
        )
        
        # Extract the full generated text
        full_text = result[0]['generated_text']
        
        # Parse the text to extract ONLY the newly generated chatbot response
        # We slice off the original prompt and grab the text up to the next newline
        new_response = full_text[len(prompt):].split('\n')[0].strip()
        
        # Display output
        print(f"Chatbot: {new_response}")

# Execute the chatbot
if __name__ == "__main__":
    run_chatbot()
```
