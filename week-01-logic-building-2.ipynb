```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
h_salary = 0
h_name = ""

for name,salary in employees.items():
    if salary>h_salary:
        h_salary = salary
        h_name = name
print(h_name ,"has a salary of:",h_salary)
```

```python
numbers = [45, 22, 89, 10, 66]
mini = numbers[0]
maxi = numbers[0]
for number in numbers:
    if number > maxi:
        maxi = number
    if number < mini:
        mini = number
print("List:",numbers)
print("max:",maxi)
print("min:",mini)
```

```python
attendance = ["P", "P", "A", "P", "P"]
head_count = 0
for attend in attendance:
    if attend == "P":
        head_count += 1
total_days = len(attendance)
percent = (head_count / total_days) * 100
print("Attendance Percentage:", percent)
```
