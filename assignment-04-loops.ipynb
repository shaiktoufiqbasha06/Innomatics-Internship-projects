```python
#inventory reorder alert
```

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")
products = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 18
}
check_inventory(products)
```

```python
#student result processing system
```

```python
def check_result(marks):
    total = 0
    for mark in marks:
        total = total + mark
    average = total / len(marks)
    print("Average:", average)
    if average >= 50:
        return "Pass"
    else:
        return "Fail"
marks = [60, 40, 70, 50]
result = check_result(marks)
print(result)
```

```python
#cab fare estimator with a retry option
```

```python
def calculate_fare(distance, peak):
    fare = 50 + (distance * 12)
    if peak == "yes":
        fare = fare + (fare * 0.25)
    return fare
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")
    fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", fare)
    retry = input("Retry? (yes/no): ")
    if retry == "no":
        break
```

```python
#employee attendance eligibility checker
```

```python
def check_attnd(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present = present + 1
    percentage = (present / len(attendance)) * 100
    print("Attendance %:", percentage)
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance = ["P", "P", "A", "P", "P"]
result = check_attnd(attendance)
print(result)
```

```python
#password strength checker
```

```python
def check_password(pswd):
    if len(pswd) < 8:
        return "Weak Password"    
    digit = False
    special = False
    special_chars = "@#$"
    for char in pswd:
        if char.isdigit():
            digit = True
        if char in special_chars:
            special = True
    if digit and special:
        return "Strong Password"
    else:
        return "Weak Password"
pswd = input("Enter password: ")
print(check_password(pswd))
```
