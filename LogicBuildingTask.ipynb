```python
#part-02 Pass / Fail Analyzer
```

```python
marks = [45,78,90,33,60]

pass_counter = 0
fail_counter = 0

for mark in marks:
  if mark >=50:
    pass_counter += 1
  else:
    fail_counter += 1
print("FINAL RESULT :-")
print("Total number of Students:", len(marks))
print("Number of passed Students:",pass_counter)
print("Number of failed Students:",fail_counter)
```

```python
#part-03 Simple Data Cleaner
```

```python
names =["Alice","bob","CHARLIE"]

cleaned_names = []

for name in names:
  cleaned_name = name.strip().lower()
  cleaned_names.append(cleaned_name)
print("Previous Names List",names)
print("Cleaned Names List",cleaned_names)
```

```python
#part-04 Message Length Analyzer
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]


for message in messages:
    length = len(message)
    print(f"Message: '{message}'")
    print(f"Length: {length}")

    if length > 10:
        print("Status: RED FLAG!! Longer than 10 characters")
    else:
        print("Status: Within the limit")

    print()
```

```python
#part-05 Error Message Detector
```

```python
errors = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for error in errors:
    if error == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
