```python
#Smart Light Controller
class controller:
    def __init__(self,light_name,action):
        self.light_name = light_name
        self.action = "OFF"  

    def light_on(self):
        self.action = "ON"

    def light_off(self):
        self.action = "OFF"

    def light_room(self):
        print(f"{self.light_name} is {self.action}")
light = controller("Bedroom Light","ON")
light.light_on()
light.light_room()
```

```python
# Employee ID Card System
class Employee_ID_reader:
    
    def __init__(self, name, emp_id, depart):
        self.name = name     
        self.emp_id = emp_id  
        self.depart = depart    
    
    def id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.depart)

employee = Employee_ID_reader("Rahul", "EMP102", "IT")
employee.id_card()
```

```python
#Mobile Contact Record
class Contact:
    
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

contact1 = Contact("Anita", "9876543210")

contact1.display_contact()
```

```python
class Product:
    
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def print_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))

item = Product("Headphones", 2499)

item.print_price_tag()
```

```python
class Movie:
    
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")


movie1 = Movie("Inception", 4.8)

movie1.display_details()
```

```python
class Delivery:
    
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    def print_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)
    
order = Delivery("Suman", "Hyderabad")

order.print_details()
```
